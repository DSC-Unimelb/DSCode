```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diamond  = sns.load_dataset("iris")
```

```python
type(diamond)
```

```python
diamond.head()
```

```python
sns.distplot(diamond['depth'])
```

```python
sns.lineplot(x = 'carat', y = 'price', data = diamond)
```

```python
sns.scatterplot(x = 'carat', y = 'price', data = diamond)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
sns.scatterplot(x = 'carat',y = 'price', hue = "clarity", data = diamond)

plt.xlabel("CARAT")
plt.ylabel("PRICE")

fig.suptitle("TITLE", fontsize = 11)
```

```python
correlation = diamond.corr()
correlation
```

```python
sns.heatmap(correlation)
```
